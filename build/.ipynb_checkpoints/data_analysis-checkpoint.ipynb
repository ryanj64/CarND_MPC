```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
data_x = []
data_y = []
data_psi = []
data_v = []
data_cte = []
data_epsi = []
data_delta = []
data_a = []

with open('data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(reader)
    for row in reader:
        #print(row)
        data_x.append( float(row[0]) )
        data_y.append( float(row[1]) )
        data_psi.append( float(row[2]) )
        data_v.append( float(row[3]) )
        data_cte.append( float(row[4]) )
        data_epsi.append( float(row[5]) )
        data_delta.append( float(row[6]) )
        data_a.append( float(row[7]) )
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_x)),data_x) #, np.arange(len(data_v)), data_v
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_y)),data_y)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_psi)),data_psi)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_v)),data_v) #, np.arange(len(data_v)), data_v
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_cte)),data_cte)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_epsi)),data_epsi) #, np.arange(len(data_v)), data_v
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_delta)),data_delta)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(16,9))
ax.plot(np.arange(len(data_a)),data_a)
plt.show()
```
